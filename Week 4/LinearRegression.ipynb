```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn import linear_model

import matplotlib.pyplot as plt
```

```python
data_url = "https://lib.stat.cmu.edu/datasets/boston"
# df = pd.read_csv(data_url, sep="\s+", skiprows = 22, header = None)  # Split on space(s)
df = pd.read_csv(data_url, sep="\\s+", skiprows = 22, header = None)  # Split on space(s)

print(df.head())
# print("x: \n", df.values[::2, :])
# print("y: \n", df.values[::2, 1])
# print("y: \n", df.values[1::2, 2])

x = np.hstack((df.values[::2, :], df.values[1::2, :2]))
y = df.values[1::2, 2]

print("Data: \n", df.head())
print("X: \n", x)
print("Y: \n", y)
```

```python
x_train, y_train, x_test, y_test = train_test_split(x, y, test_size=0.8, random_state=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8, random_state=1)
print("x_train\n", x_train)
print("y_train\n", y_train)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x_train, y_train)

print("Coefficients: ", reg.coef_)
print("Variance score: ", reg.score(x_test, y_test))
```
